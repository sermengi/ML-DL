```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
observations = 1000
xs = np.random.uniform(-10,10,(observations,1))
zs = np.random.uniform(-10,10,(observations,1))

inputs = np.column_stack((xs,zs))
```

```python
noise = np.random.uniform(-1,1,(observations,1))
targets = 2*xs - 3*zs + 5 + noise
```

```python
targets_plot = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot(xs,zs,targets_plot)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')
ax.view_init(azim=100)
plt.show()
```

```python
init_range = 0.1
weights = np.random.uniform(-init_range,init_range,(2,1))
bias = np.random.uniform(-init_range,init_range,1)
learning_rate = 0.02
```

```python
loss_func = []
for i in range(100):
    outputs = np.dot(inputs,weights)+bias
    
    deltas = outputs - targets
    loss = np.sum(deltas**2) / 2 / observations
    
    deltas_scaled = deltas / observations
    weights = weights - learning_rate*np.dot(inputs.T,deltas_scaled)
    bias = bias - learning_rate*np.sum(deltas_scaled)
    loss_func.append(loss)
```

```python
iteration = np.array(range(1,101))
plt.plot(iteration,loss_func)
plt.xlabel('Iteration',fontsize=15)
plt.ylabel('Loss Function',fontsize=15)
plt.show()
```
